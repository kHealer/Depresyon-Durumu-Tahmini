```python
from sklearn.model_selection import train_test_split, GridSearchCV,cross_val_score
from sklearn.metrics import mean_squared_error,r2_score
from sklearn import model_selection
from sklearn.neighbors import KNeighborsRegressor
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df.info()
```

```python
df.melanch.fillna(2.0, inplace = True)
```

```python
list(df.columns)
```

```python
for c in df:
    df[c] = df[c].astype("category").cat.codes
df
```

```python
X = df.drop(['MadrsMean'], axis=1)
y = df['MadrsMean']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 42)
```

```python
knn_model = KNeighborsRegressor()
```

```python
model = knn_model.fit(X_train, y_train)
```

```python
new_info = [[3],[1],[3],[1],[1],[1],[3],[1],[1],[4],[6],[5]]
```

```python
new_info = pd.DataFrame(new_info).T
```

```python
model.predict(new_info)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

# Test error

```python
mean_squared_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

# Train error

```python
mean_squared_error(y_train, model.predict(X_train))
```

```python
np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
```

# train error different k values (not validated)

```python
for k in range(18):
    k=k+1
    k_model = KNeighborsRegressor(n_neighbors = k).fit(X_train, y_train)
    y_pred = k_model.predict(X_train)
    error = np.sqrt(mean_squared_error(y_train, y_pred))
    print('k value = ' , k , 'error value =' , error)
```

# test error different k values (not validated)

```python
for k in range(18):
    k=k+1
    k_model = KNeighborsRegressor(n_neighbors = k).fit(X_train, y_train)
    y_pred = k_model.predict(X_test)
    error = np.sqrt(mean_squared_error(y_test, y_pred))
    print('k value = ' , k , 'error value =' , error)
```
